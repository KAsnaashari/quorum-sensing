```python
from sklearn.preprocessing import StandardScaler

features = ['COM_x', 'COM_y', 'COM_z']
x = dock_data.loc[:, features].values
y = dock_data.loc[:,['score']].values
scaler = StandardScaler()
x = scaler.fit_transform(x)
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=3)
principalComponents = pca.fit_transform(x)
principalDf = pd.DataFrame(data = principalComponents, columns = ['pc1', 'pc2', 'pc3'])
print(principalDf.size)
principalDf.head()
```

```python
finalDf = pd.concat([principalDf, dock_data[['score']]], axis = 1)
pca.explained_variance_ratio_
```

```python
finalDf.sort_values(by='score').head()
```

```python
fig = px.scatter_3d(finalDf, x='pc1', y ='pc2', z='pc3', color='score')
fig.show()
```

```python
fig = px.scatter(finalDf, x='pc1', y='pc2', color='score')
fig.show()
```

```python
from sklearn.cluster import KMeans

ks = range(1, 10)
inertias = []
for k in ks:
    # Create a KMeans instance with k clusters: model
    model = KMeans(n_clusters=k)
    
    # Fit model to samples
    model.fit(principalDf.iloc[:,:3])
    
    # Append the inertia to the list of inertias
    inertias.append(model.inertia_)
    
plt.plot(ks, inertias, '-o', color='black')
plt.xlabel('number of clusters, k')
plt.ylabel('inertia')
plt.xticks(ks)
plt.show()
```

```python
k = 2

kmeans2 = KMeans(n_clusters=k)
    
# Fit model to samples
y_kmeans2 = kmeans2.fit_predict(principalDf.iloc[:,:3])

kmeans2_centroids = kmeans2.cluster_centers_
print(kmeans2_centroids)
```

```python
plt.scatter(principalDf['pc1'], principalDf['pc2'], c= kmeans2.labels_.astype(float), s=50, alpha=0.5)
plt.scatter(kmeans2_centroids[:, 0], kmeans2_centroids[:, 1], c='red', s=50)
```

```python
k = 3

kmeans3 = KMeans(n_clusters=k)
    
# Fit model to samples
y_kmeans3 = kmeans3.fit_predict(principalDf.iloc[:,:3])

kmeans3_centroids = kmeans3.cluster_centers_
print(kmeans3_centroids)
```

```python
plt.scatter(principalDf['pc1'], principalDf['pc2'], c= kmeans3.labels_.astype(float), s=50, alpha=0.5)
plt.scatter(kmeans3_centroids[:, 0], kmeans3_centroids[:, 1], c='red', s=50)
```

```python
k = 4

kmeans4 = KMeans(n_clusters=k)
    
# Fit model to samples
y_kmeans4 = kmeans4.fit_predict(principalDf.iloc[:,:3])

kmeans4_centroids = kmeans4.cluster_centers_
print(kmeans4_centroids)
```

```python
plt.scatter(principalDf['pc1'], principalDf['pc2'], c= kmeans4.labels_.astype(float), s=50, alpha=0.5)
plt.scatter(kmeans4_centroids[:, 0], kmeans4_centroids[:, 1], c='red', s=50)
```

```python
dock_data['kmeans2'] = y_kmeans2
dock_data['kmeans3'] = y_kmeans3
dock_data['kmeans4'] = y_kmeans4
dock_data.head()
```

```python
kmeans2_centroid_inverse = pca.inverse_transform(kmeans2_centroids)
kmeans2_centroid_inverse = scaler.inverse_transform(kmeans2_centroid_inverse)
print(list(enumerate(kmeans2_centroid_inverse)))
```

```python
kmeans3_centroid_inverse = pca.inverse_transform(kmeans3_centroids)
kmeans3_centroid_inverse = scaler.inverse_transform(kmeans3_centroid_inverse)
print(list(enumerate(kmeans3_centroid_inverse)))
```

```python
kmeans4_centroid_inverse = pca.inverse_transform(kmeans4_centroids)
kmeans4_centroid_inverse = scaler.inverse_transform(kmeans4_centroid_inverse)
print(list(enumerate(kmeans4_centroid_inverse)))
```
